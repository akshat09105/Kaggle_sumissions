```python
mydataset=pd.read_csv("train.csv")
```

```python
from pycaret.regression import *
setup(data=mydataset,target='output',normalize=True,normalize_method="minmax",transformation=True,remove_outliers=True,train_size=0.9)#pca#transformation=True
```

```python
cm=compare_models()
```

```python
mydataset.corr()['output'].sort_values(ascending=False)
```

```python
lightgbmModel=create_model("gbr")
```

```python
mynewdata=pd.read_csv("test.csv")
```

```python
predictions=predict_model(lightgbmModel,data=mynewdata)
predictions
```

Tuning the Model

```python
tunegbbrModel=tune_model(lightgbmModel,n_iter=7)
predictions=predict_model(tunegbbrModel,data=mynewdata)
predictions
```

```python
dictionary=pd.DataFrame({'id':predictions['id'],'output':predictions['prediction_label']})
dictionary
```

```python
dictionary.to_csv("Answer.csv",index=False)
```
